## Import the playlist csv: moosic_df

```python
import pandas as pd

# Read the CSV as a Pandas dataframe
csv_url = 'https://docs.google.com/spreadsheets/d/1vPpAYd7yLknpzuvFtJLv0K7ZJu1X8GR8G_aoVjDA31c/export?format=csv&id=1vPpAYd7yLknpzuvFtJLv0K7ZJu1X8GR8G_aoVjDA31c&gid=1481867361'
moosic_df = pd.read_csv(csv_url)

# Display the dataframe
moosic_df.head()

```

```python
# Create a new dataframe for playlists
playlists_df = moosic_df.copy()

# Group songs by cluster and calculate the average value for each group
grouped_df = playlists_df.groupby('cluster').mean().reset_index()

# Display the grouped dataframe
grouped_df.head()

```

## Import the df 5000 csv

```python
# Read the CSV as a Pandas dataframe
sheet_id = "1tfnfAcZq8GwFCmL4cyaflP15iY09XEXwvxSL5MwCGVQ"
sheet_name = "466517768"
csv_url = f"https://docs.google.com/spreadsheets/d/{sheet_id}/export?format=csv&gid={sheet_name}"
songs_df = pd.read_csv(csv_url)

# Display the dataframe
songs_df.head()
```

## Join both DFs

```python
# Add an 'id' column as the index for moosic_df
moosic_df['id'] = moosic_df.index

# Add an 'id' column as the index for songs_df
songs_df['id'] = songs_df.index

# Merge the dataframes based on the common column 'id'
merged_df = songs_df.merge(moosic_df[['id', 'cluster']], on='id', how='left')

# Display the merged dataframe
merged_df.head()
```

## Group by 'cluster'

```python
# Create a new dataframe grouped by 'cluster'
grouped_df = merged_df.groupby('cluster').apply(lambda x: x[['name', 'artist']].reset_index(drop=True))

# Reset the index of the grouped dataframe
grouped_df = grouped_df.reset_index(level='cluster')

# Rename the columns for clarity
grouped_df.columns = ['cluster', 'song', 'artist']

# Display the grouped dataframe
grouped_df

```

## Locate only relevant groups: from 6 to 10

```python
# Filter the grouped dataframe for clusters 6 to 10
filtered_df = grouped_df[grouped_df['cluster'].between(6, 10)]

# Display the filtered dataframe
filtered_df
```

## Let's make seperate playlists

```python
# Filter the grouped dataframe for clusters 6 to 10
filtered_df = grouped_df[grouped_df['cluster'].between(6, 10)]

# Create a dictionary to store the dataframes for each cluster
cluster_dfs = {}

# Iterate over each cluster in the filtered dataframe
for cluster in filtered_df['cluster'].unique():
    # Select the songs and artists for the current cluster
    songs_artists = filtered_df.loc[filtered_df['cluster'] == cluster, ['song', 'artist']]

    # Create a new dataframe for the current cluster
    cluster_df = pd.DataFrame(songs_artists)

    # Add the cluster dataframe to the dictionary with cluster number as the key
    cluster_dfs[cluster] = cluster_df

# Accessing the dataframe for cluster 6
cluster_6_df = cluster_dfs[6]

# Accessing the dataframe for cluster 7
cluster_7_df = cluster_dfs[7]

# Accessing the dataframe for cluster 8
cluster_8_df = cluster_dfs[8]

# Accessing the dataframe for cluster 9
cluster_9_df = cluster_dfs[9]

# Accessing the dataframe for cluster 10
cluster_10_df = cluster_dfs[10]
```

```python
# Create an empty dictionary to store the concatenated results
concatenated_data = {}

# Iterate over each cluster DataFrame
for cluster_num, df in cluster_dfs.items():
    # Concatenate the 'song' and 'artist' columns with a '|'
    df['sing'] = df['song'] + ' | ' + df['artist']
    # Store the concatenated column in the dictionary
    concatenated_data[cluster_num] = df['sing']

# Print the concatenated results for each cluster
for cluster_num, songs in concatenated_data.items():
    print(f"Cluster {cluster_num}:\n{songs}\n")

```
